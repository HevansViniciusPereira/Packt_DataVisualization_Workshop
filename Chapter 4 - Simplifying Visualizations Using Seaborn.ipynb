# Comparing IQ Scores for Different Test Groups by Using a Box Plot

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
data = pd.read_csv('Datasets/iq_scores.csv')
data.head()
```

```python
# Access the data of each test group in the column
group_a = data[data.columns[0]].tolist()
group_b = data[data.columns[1]].tolist()
group_c = data[data.columns[2]].tolist()
group_d = data[data.columns[3]].tolist()
```

```python
# Print the values of each group
print(group_a)
```

```python
print(group_b)
```

```python
print(group_c)
```

```python
print(group_d)
```

```python
# construct a DataFrame from this data
df = pd.DataFrame({'Groups': ['Group A'] * len(group_a) + 
                  ['Group B'] * len(group_b) +
                  ['Group C'] * len(group_c) + 
                  ['Group D'] * len(group_d),
                  'IQ scores': group_a + group_b + group_c + group_d})
df.head()
```

```python
# create a box plot
plt.figure(dpi=150)
sns.set_style('whitegrid')
sns.boxplot('Groups', 'IQ scores', data=df)
sns.despine(left=True, right=True, top=True)
plt.title('IQ scores for different test groups')
plt.show()
```

# Surface Temperature Analysis

```python
# import packages
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('Datasets/northern_surface_temperature.csv', index_col=['Year']).transpose()
data
```

```python
# Create a custom-diverging palette that diverges to blue for low values and to red for high values
heat_colormap = sns.diverging_palette(240, 15, s=99, as_cmap=True)
```

```python
# Plot the heatmap for every 5 years
plt.figure(dpi=200)
sns.heatmap(data.iloc[:, ::5], cmap=heat_colormap, center=0)
plt.title('Temperature changes from 1880 to 2015 (base period 1951-1980)')
plt.savefig('temperature_change.png', dpi=300, bbox_inches='tight')
```

# Water Usage Revisited

```python
# import package
import squarify
```

```python
# laod the data
df = pd.read_csv('Datasets/water_usage.csv', index_col=0)
df
```

```python
# Create a list of labels by accessing each column from the preceding dataset
labels = df['Usage'] + ' (' + df['Percentage'].astype('str') + '%)'
labels
```

```python
# create a tree map visualization
plt.figure(dpi=200)
squarify.plot(sizes=df['Percentage'],
             label=labels,
             color=sns.light_palette('green', df.shape[0]))
plt.axis('off')
plt.title('Water Usage')
plt.show()
```
